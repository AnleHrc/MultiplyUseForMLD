```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('output.csv', index_col=0)
data.head()
sns.heatmap(data, cmap='coolwarm')
plt.show()
```

```python
# 针对每一列数据画直方图
for i in range(1, 11):
    plt.figure()
    sns.histplot(data.iloc[:, i-1], kde=True)
    plt.title(f"Distribution of Threshold {i}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
```

```python
plt.figure()
sns.histplot(data, kde=True, multiple="stack")
plt.title("Distribution of All Thresholds")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend(title="Threshold", labels=["Threshold 1", "Threshold 2", "Threshold 3", "Threshold 4", "Threshold 5", "Threshold 6", "Threshold 7", "Threshold 8", "Threshold 9", "Threshold 10"])
plt.show()
```

```python
# Reshape data to long format
data = data.unstack().reset_index(name='value')
data = data.rename(columns={'level_0': 'threshold', 'level_1': 'id'})

# Create FacetGrid
g = sns.FacetGrid(data, col='threshold', col_wrap=5, sharex=False, sharey=False)
g.map(sns.histplot, 'value', kde=True)
g.set_titles(col_template="Threshold {col_name}")

plt.subplots_adjust(top=0.9)
g.fig.suptitle('Distribution of Threshold Values by Rule ID', fontsize=16)

plt.show()
```

```python
plt.figure()
sns.histplot(data, kde=True, multiple="stack")
plt.title("Distribution of All Thresholds")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend(title="Threshold", labels=["Threshold 1", "Threshold 2", "Threshold 3", "Threshold 4", "Threshold 5", "Threshold 6", "Threshold 7", "Threshold 8", "Threshold 9", "Threshold 10"])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('output.csv', index_col=0)
data.head()

# 创建3D图形对象
fig = plt.figure()
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')

# 生成x、y坐标
x = [i+1 for i in range(len(data.columns))]
y = [i+1 for i in range(len(data.index))]

# 将数据中的每个值作为z值
X, Y = [], []
for i in range(len(y)):
    for j in range(len(x)):
        X.append(x[j])
        Y.append(y[i])
        ax.scatter(x[j], y[i], data.iloc[i, j], c='b')

# 设置坐标轴标签
ax.set_xlabel('Threshold')
ax.set_ylabel('Rule')
ax.set_zlabel('Value')

# 显示图形
plt.show()
```

```python


# 获取数据的坐标
x, y, z = [], [], []
for i in range(len(data)):
    for j in range(len(data.columns)):
        x.append(j+1)
        y.append(i+1)
        z.append(data.iloc[i,j])

# 绘制散点图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
ax.set_xlabel('Threshold')
ax.set_ylabel('Rule ID')
ax.set_zlabel('Value')
plt.show()
```

```python


# 绘制柱状图
fig = plt.figure()
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')
xpos = range(len(data.columns))
ypos = range(len(data))
xpos, ypos = np.meshgrid(xpos, ypos)
xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like(xpos)
dx = 0.5 * np.ones_like(zpos)
dy = dx.copy()
dz = data.values.flatten()
ax.bar3d(xpos, ypos, zpos, dx, dy, dz)
ax.set_xlabel('Threshold')
ax.set_ylabel('Rule ID')
ax.set_zlabel('Value')
plt.show()
```
