```python
import pandas as pd
import numpy as np
import itertools

df = pd.read_pickle('Data/data_a2.pkl')
df.head(10)

```

```python
# 变量
m = 1000000
p = 0.08

def cal_w2(lis): # 计算三张卡组合起来的利润,x是[t1,t2,t3],y是[h1,h2,h3]
    x = lis[3]*lis[4]*lis[5]
    y = np.mean(lis[6:])
    return m*x*(p-(1+p)*y)


# 2
Data2 = []
for i, j, k in itertools.product(range(10), range(10), range(10)):
    t1,h1 = df.iloc[i,0]
    t2,h2 = df.iloc[j,1]
    t3,h3 = df.iloc[k,2]
    Data2.append([i,j,k,t1,t2,t3,h1,h2,h3])

res2 = pd.DataFrame(Data2,columns=['Threshold1','Threshold2','Threshold3','t1','t2','t3','h1','h2','h3'])
res2['w'] = res2.apply(cal_w2,axis=1)
res2 = res2.sort_values('w',ascending=0)
res2.head(5)

print(res2[0:5])
# 3
dd = df.copy()
dd = dd.applymap(lambda x:np.array(x))
dd = dd.to_numpy()
```
